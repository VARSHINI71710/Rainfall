```python
import matplotlib.pyplot as plt

skew_values = num_df.skew(numeric_only=True)

plt.figure(figsize=(10,6))
skew_values.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Skewness of Numeric Features", fontsize=14)
plt.xlabel("Features", fontsize=12)
plt.ylabel("Skewness", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.axhline(y=0, color='red', linestyle='--') 
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

num_df = df_random.select_dtypes(include=['int64', 'float64']).copy()

cols_to_square = ['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation',
                  'Sunshine', 'WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm',
                  'Humidity9am', 'Humidity3pm', 'Pressure9am', 'Cloud9am', 'Temp9am']

for col in cols_to_square:
    if col in num_df.columns:
        num_df[col] = num_df[col] ** 2

skew_values = num_df.skew(numeric_only=True)

plt.figure(figsize=(10,6))
skew_values.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title("Skewness of Selected Numeric Columns", fontsize=14)
plt.xlabel("Features", fontsize=12)
plt.ylabel("Skewness", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.axhline(y=0, color='red', linestyle='--')  
plt.tight_layout()
plt.show()
```
